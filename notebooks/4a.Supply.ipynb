# **Bitcoin: A Deflationary Asset with Increasing Demand**

## *Introduction to Bitcoin Supply Dynamics*

### Overview

Bitcoin (BTC) distinguishes itself in the realm of assets due to its unique supply characteristics. Unlike traditional fiat currencies or the majority of other cryptocurrencies, Bitcoin possesses a fixed maximum supply and a mechanism that progressively reduces the rate of new coin issuance over time. These features render Bitcoin a deflationary asset, standing in stark contrast to inflationary fiat currencies. This notebook focuses on the key aspects of Bitcoin's supply dynamics and explicates why they contribute to its potential value as a digital asset.

## Key Characteristics of Bitcoin's Supply

### *1. Fixed Maximum Supply*

Bitcoin has an immutable, hard-coded maximum supply cap of 21 million coins. This limit is an integral and unchangeable part of the Bitcoin protocol, guaranteeing that no more than 21 million BTC will ever be created. This scarcity is fundamental to Bitcoin's value proposition and economic model.

### *2. Decreasing Issuance Rate: The Halving*

The rate at which new bitcoins are created is not constant. Instead, it adheres to a predetermined schedule known as the "halving." Approximately every four years (or more precisely, every 210,000 blocks), the number of new bitcoins generated per block is reduced by half. This mechanism systematically diminishes the rate of new bitcoin introduction into circulation.

The halving schedule is as follows:
- 2009-2012: 50 BTC per block
- 2012-2016: 25 BTC per block
- 2016-2020: 12.5 BTC per block
- 2020-2024: 6.25 BTC per block
- 2024-2028: 3.125 BTC per block
- And so forth...

This decreasing issuance rate engenders a disinflationary effect, gradually reducing Bitcoin's inflation rate over time.

### *3. Lost Coins*

Over the course of time, a certain percentage of bitcoins are lost due to various reasons, including forgotten private keys, accidental loss, or the death of the owner without passing on access information. These lost coins effectively reduce the circulating supply, potentially enhancing the scarcity of available bitcoins.

### How Coins Can Be Lost

### *1. Loss of Private Keys*

Bitcoin ownership is determined by private keys. If the private key to a Bitcoin address is lost, the coins at that address become inaccessible and are effectively removed from circulation. 

- **Example**: An early Bitcoin adopter stored their private keys on a hard drive, which was later discarded by mistake. The hard drive ended up in a landfill, rendering the Bitcoins permanently inaccessible.
  
### *2. Death Without Passing On Keys*

If a Bitcoin holder dies without passing on their private keys or information on how to access their Bitcoin wallet to heirs or executors, the coins are lost in perpetuity.

- **Example**: A Bitcoin investor with a substantial amount of BTC passes away unexpectedly. Their family is unaware of how to access the digital wallet, leading to a permanent loss of those coins.

### *3. Forgotten Wallet Passwords*

Some Bitcoin wallets are encrypted with passwords. If the password is forgotten and no backup exists, the wallet cannot be accessed, resulting in the loss of any Bitcoin it contains.

- **Example**: A user sets up a secure Bitcoin wallet with a complex password but fails to record the password anywhere. Years later, the user cannot recall the password and thus cannot access the wallet.

### *4. Software or Hardware Failure*

Bitcoin wallets can be stored on various devices such as computers, USB drives, or specialized hardware wallets. If these devices fail and there is no backup, the Bitcoin can be lost.

- **Example**: A hardware wallet malfunctions, and the owner did not create a recovery phrase or backup. The Bitcoins stored on the hardware wallet are lost.

### *5. Sending to Incorrect Addresses*

Bitcoin transactions are irreversible. If Bitcoin is sent to an incorrect or non-existent address, it cannot be retrieved.

- **Example**: An individual intends to transfer Bitcoin to a friend but accidentally enters a wrong address. The Bitcoin is sent to a non-existent or incorrect address, making it irretrievable.

## The Deflationary Nature of Bitcoin

Bitcoin's deflationary nature stems from the combination of its fixed supply cap, decreasing issuance rate, and the phenomenon of lost coins. As the supply growth decelerates and eventually ceases, coupled with the continuous loss of coins, the available supply may actually decrease over time. This contrasts sharply with fiat currencies, where central banks can increase the money supply at will, often leading to inflation.

## Implications of Bitcoin's Supply Dynamics

1. **Scarcity**: The fixed supply cap and decreasing issuance rate contribute to Bitcoin's digital scarcity, which is often compared to precious metals like gold.

2. **Predictability**: Unlike fiat currencies, Bitcoin's supply schedule is highly predictable and can be accurately forecasted decades into the future.

3. **Store of Value Potential**: The deflationary nature of Bitcoin supports its narrative as a potential store of value, especially in environments of high inflation.

4. **Economic Incentives**: The halving events create unique economic incentives for miners and investors, potentially influencing Bitcoin's market dynamics.

# **1. Bitcoin Available Supply Model and Visualization**

## Introduction

While Bitcoin's total supply is capped at 21 million coins, the actual number of bitcoins available for circulation is likely lower due to various factors, primarily the permanent loss of access to some coins. This section introduces a model for estimating the available supply of Bitcoin and presents a visualization of this estimate alongside the total supply.

## The Available Supply Model

Our model for estimating the available supply of Bitcoin takes into account the phenomenon of lost coins. Research indicates that approximately 4% of available bitcoins are lost each year. This loss rate is not constant but is expected to decrease over time as users become more educated about proper storage and backup procedures.

### Key Components of the Model

1. **Initial Loss Rate**: We start with an initial annual loss rate of 2.5% (0.025).
2. **Decay Rate**: The loss rate is assumed to decay by 1.5% (0.985) each year, reflecting improving security practices over time.
3. **Cumulative Loss**: The model calculates the cumulative effect of these losses over time.

### Mathematical Representation

The percentage of coins lost over time can be represented as:

```
lost_coins_percentage = 1 - exp(-initial_loss_rate * (decay_rate^years) * years)
```

Where:
- `initial_loss_rate` is the starting annual rate of coin loss (0.025 in our model)
- `decay_rate` is the annual reduction in the loss rate (0.985 in our model)
- `years` is the number of years since the start of Bitcoin

## Visualization

The `plot_available_supply` function creates a visualization that illustrates the relationship between the total Bitcoin supply and the estimated available supply based on our model.

### Key Elements of the Plot

1. **BTC Supply - Forecast (Orange Line)**: This line represents the projected total supply of Bitcoin over time, including future dates based on the known issuance schedule.

2. **BTC Supply - Historical (Blue Line)**: This line shows the actual historical data of Bitcoin's total supply.

3. **BTC Supply Available - Forecast (Green Line)**: This line represents our model's estimate of the available supply of Bitcoin, taking into account the estimated loss of coins over time.

4. **Halving Date Markers**: Vertical lines mark the dates of past and future halving events, highlighting these significant points in Bitcoin's supply schedule.

5. **Maximum Supply Line**: A horizontal red dashed line at 21 million BTC represents the maximum supply cap.

6. **Lost Coins Percentage**: A text box on the plot displays the estimated percentage of coins lost based on the most recent data point.

### Interpretation of the Visualization

- The divergence between the total supply (orange/blue lines) and the available supply (green line) illustrates the cumulative effect of lost coins over time.
- The stepwise pattern in the supply lines reflects the impact of halving events on the rate of new coin issuance.
- The flattening of the supply curves as they approach the 21 million cap demonstrates Bitcoin's disinflationary nature.
- The growing gap between total and available supply highlights the increasing scarcity of accessible bitcoins.

## Implications of the Available Supply Model

1. **Increased Scarcity**: The model suggests that the effective circulating supply of Bitcoin may be significantly lower than the total coins mined, potentially increasing its scarcity and, by extension, its value proposition.

2. **Deflationary Pressure**: As the rate of coin loss potentially outpaces the rate of new coin issuance, especially after future halvings, it could create deflationary pressure on the Bitcoin economy.

3. **Long-term Holder Incentives**: The model underscores the potential benefit of long-term holding and proper security practices, as lost coins effectively redistribute value to the remaining accessible supply.

4. **Market Dynamics**: Understanding the available supply can provide insights into potential market dynamics, as it more accurately represents the coins that can actively participate in the economy.

# **2. Comparison of Bitcoin Supply and M2 Money Supply**

## Introduction

To appreciate the unique characteristics of Bitcoin's supply dynamics, it's instructive to compare it with traditional monetary measures. This section focuses on a comparison between Bitcoin's supply and the M2 money supply, a key measure of money in the traditional financial system. 

## Understanding M2 Money Supply

Before diving into the comparison, it's important to understand what M2 money supply represents:

1. **Definition**: M2 is a measure of the money supply that includes cash, checking deposits, and easily convertible near money. It's broader than M1 and is used by economists to quantify the amount of money in circulation and to explain different economic monetary conditions.

2. **Components**: M2 typically includes:
   - All elements of M1 (cash and checking deposits)
   - Savings deposits
   - Money market securities
   - Mutual funds
   - Other time deposits

3. **Control**: Unlike Bitcoin, the M2 money supply is controlled by central banks and can be influenced by monetary policy decisions.

## Visualization

The `plot_m2_supply` function creates a visualization that compares the supply of Bitcoin with the M2 money supply over time.

### Key Elements of the Plot

1. **BTC Supply - Forecast (Orange Line)**: Projected total supply of Bitcoin over time.

2. **BTC Supply - Historical (Blue Line)**: Actual historical data of Bitcoin's total supply.

3. **M2 Money Supply (Purple Line, Secondary Y-axis)**: Historical data of the U.S. M2 money supply.

4. **Halving Date Markers**: Vertical lines marking Bitcoin halving events.

5. **Maximum Bitcoin Supply Line**: Horizontal red dashed line at 21 million BTC.

### Interpretation of the Visualization

- **Supply Growth Patterns**: The Bitcoin supply curve shows a predictable, diminishing growth rate, while the M2 supply typically shows a more consistent or even accelerating growth rate.

- **Supply Caps**: Bitcoin's supply asymptotically approaches its 21 million cap, while the M2 supply has no inherent upper limit.

- **Influence of Economic Events**: Major economic events (like recessions or crises) often correspond with accelerated growth in the M2 supply, while Bitcoin's supply remains unaffected by such events.

- **Predictability**: Bitcoin's future supply can be accurately predicted, while future M2 supply is subject to policy decisions and economic conditions.

## Key Comparisons

1. **Inflation vs. Deflation**:
   - M2: Generally inflationary, with the money supply typically increasing over time.
   - Bitcoin: Disinflationary, with a decreasing rate of new coin issuance, potentially becoming deflationary when accounting for lost coins.

2. **Supply Control**:
   - M2: Centrally controlled by monetary authorities.
   - Bitcoin: Decentralized, controlled by its protocol.

3. **Predictability**:
   - M2: Subject to policy decisions and economic conditions, making long-term prediction challenging.
   - Bitcoin: Highly predictable due to its predetermined issuance schedule.

4. **Response to Economic Conditions**:
   - M2: Often expanded during economic downturns as a policy tool.
   - Bitcoin: Unaffected by economic conditions, maintaining its predetermined issuance rate.

5. **Transparency**:
   - M2: Calculation methodologies can change, and some components may not be immediately visible.
   - Bitcoin: Fully transparent, with every transaction publicly recorded on the blockchain.

## Implications of the Comparison

1. **Monetary Policy**: The comparison highlights the fundamental differences in monetary policy between traditional fiat systems and Bitcoin's protocol-based system.

2. **Store of Value**: Bitcoin's predictable and capped supply contrasts with the potentially unlimited growth of M2, supporting Bitcoin's narrative as a potential store of value.

3. **Economic Calculation**: The predictability of Bitcoin's supply could potentially facilitate more accurate long-term economic calculation compared to the less predictable M2 supply.

4. **Globalization of Money**: While M2 is typically a national measure, Bitcoin's borderless nature presents a new paradigm in global money supply dynamics.

5. **Crisis Response**: The comparison illustrates the different tools available for economic crisis response in traditional and Bitcoin-based systems.
