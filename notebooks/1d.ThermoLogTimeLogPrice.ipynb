# Ensemble Model for Bitcoin Price Analysis and Prediction

## Introduction

This project introduces an advanced ensemble model designed to analyze and predict Bitcoin (BTC) price movements. By integrating two distinct modeling approaches, it provides a more robust and comprehensive framework for understanding Bitcoin's price dynamics, both in current market conditions and for future projections.

## Core Components

### 1. Log-Log Model

The Log-Log model leverages a double logarithmic scale to analyze Bitcoin's price trends, making it particularly suited for assets with exponential growth patterns. This approach helps linearize Bitcoin's trend over time, making long-term behavior more interpretable.

Key features:
- Applies logarithmic scaling to both price and time axes
- Reveals long-term trends and growth rates
- Effective for assets characterized by high volatility and exponential growth

### 2. ThermoModel

The ThermoModel combines cubic and logarithmic regression techniques to analyze and predict Bitcoin price movements. By creating dynamic price bands and oscillators, it identifies potential buy and sell zones, offering a comprehensive tool for understanding market trends and informing investment decisions.

Key features:
- Uses logarithmic and cubic regressions to model Bitcoin's growth
- Identifies long-term trends and potential price shifts
- Especially effective for volatile, fast-growing assets

## Ensemble Methodology

### Data Fusion

The ensemble model begins by merging the data from both the Log-Log and Thermo models. This ensures the time series from each model is aligned, resulting in a synchronized dataset for further analysis.

### Price Corridor Calculation

A key innovation of the ensemble approach is the construction of a unified price corridor:

1. **Lower Band**: The average of the lower bounds from both models.
2. **Upper Band**: The average of the upper bounds from both models.

By combining the strengths of each model, this method offers a more reliable price range estimation.

### Oscillator Development

The model introduces a custom oscillator that measures the current price position within the defined price corridor:

- Scaled from 0 to 100, where 0 represents the lower band and 100 the upper band
- Offers a normalized indication of Bitcoin's price position, regardless of absolute price levels

### Trend Smoothing

To reduce noise and highlight key trends, the oscillator undergoes smoothing:

- Uses LOWESS (Locally Weighted Scatterplot Smoothing) for trend refinement
- Helps identify clear buy and sell zones
- Improves visibility of medium- and long-term trends

## Visualization and Analysis

The ensemble model generates a comprehensive visual representation:

### Main Price Chart

- Displays Bitcoin price on a logarithmic scale
- Shows upper and lower bands of the price corridor
- Includes annotations for current price levels and projected peaks

### Oscillator Chart

- Visualizes the smoothed oscillator
- Highlights potential buy and sell zones
- Offers a quick reference for market positioning

3. **Better Risk Management**: Clearly defined buy and sell zones support better risk assessment and portfolio management.
4. **Adaptability**: The model can adjust to changing market conditions, incorporating multiple analytical perspectives.
5. **Forward-Looking Projections**: Provides a framework for forecasting future price ranges and market cycles.
