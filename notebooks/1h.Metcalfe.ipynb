# Bitcoin Metcalfe's Law Analysis

## Goal
Quantify Bitcoin's fundamental value using network activity to identify overvaluation/undervaluation periods and develop a data-driven trading indicator based on network effects theory.

## Theoretical Background
Metcalfe's Law states that a network's value is proportional to the square of its users. For Bitcoin:
- **Network effect**: More users = exponentially higher value
- **Fundamental basis**: Price should correlate with actual network usage
- **Market inefficiency**: Price often deviates from fundamental value

## Core Formula
$$V = A(t) \times \frac{n(n-1)}{2c}$$
Where:
- $V$ = Network Value (Bitcoin market cap)
- $n$ = Network nodes (transactions or active addresses) 
- $c$ = Circulating supply
- $A(t)$ = Time-dependent affinity parameter

The formula represents actual possible network connections ($\frac{n(n-1)}{2}$) adjusted for supply dilution.

## Affinity Models
| Model | Formula | Use Case |
|-------|---------|----------|
| Constant | $A = a_1$ | Baseline model |
| Linear | $A = a_1(1 + a_2 t)$ | Steady adoption growth |
| Exponential | $A = a_1 e^{a_2 t}$ | Accelerating adoption |
| Quadratic | $A = a_1(1 + a_2 t + a_3 t^2)$ | Complex evolution patterns |
